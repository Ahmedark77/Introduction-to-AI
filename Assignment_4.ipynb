```python
# Task 9: Iterator and Generator
class EvenIterator:
    def __init__(self):
        self.current = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= 5:
            raise StopIteration
        self.count += 1
        self.current += 2
        return self.current

def even_generator():
    for i in range(2, 12, 2):
        yield i

print("Task 9:")
print("Iterator:", list(EvenIterator()))
print("Generator:", list(even_generator()))
print("\n")
```

```python
# Task 10: Map, Reduce, Filter
numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x**2, numbers))
product = reduce(lambda x, y: x * y, numbers)
evens = list(filter(lambda x: x % 2 == 0, numbers))

print("Task 10:")
print(f"Original: {numbers}")
print(f"Map (square): {squared}")
print(f"Reduce (product): {product}")
print(f"Filter (evens): {evens}")
print("\n")
```

```python
# Task 11: Rectangle Class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect1 = Rectangle(5, 3)
rect2 = Rectangle(7, 4)

print("Task 11:")
print(f"Rectangle 5x3 - Area: {rect1.area()}, Perimeter: {rect1.perimeter()}")
print(f"Rectangle 7x4 - Area: {rect2.area()}, Perimeter: {rect2.perimeter()}")
print("\n")
```

```python
# Task 12: Pandas Data Analysis
df_emp = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

print("Task 12:")
print("Original Data:")
print(df_emp)

# Average salary by department
print("\nAverage salary by department:")
print(df_emp.groupby('Department')['Salary'].mean())

# Employees with salary > 60000
high_earners = df_emp[df_emp['Salary'] > 60000]['Name']
print(f"\nEmployees with salary > 60000: {list(high_earners)}")
```
